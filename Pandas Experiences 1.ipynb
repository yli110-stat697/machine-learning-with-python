something wrong is here. If the regions are not sorted, how do you make sure the rolling mean is the desired mean of 25 points.

```python
organic_df.set_index("Date", inplace = True)
organic_df.sort_index(inplace = True)
organic_df.head(6)
```

```python
graph_df = pandas.DataFrame()

for region in organic_df["region"].unique():
    print(region)
    region_df = organic_df.copy()[organic_df["region"] == region]
    region_df.sort_index(inplace = True)
    region_df[f'{region}_price'] = region_df["AveragePrice"].rolling(25).mean()
    
    
    if graph_df.empty:
        graph_df = region_df[[f'{region}_price']] #[[]] gives you df, [] gives you series
    else:
        graph_df = graph_df.join(region_df[f'{region}_price']) #keep adding columns
        
graph_df.tail()
```

we want to put the State on top as new columns, just like we did with avocado dataset, but this time, use the pandas's way.

```python
gb = df.groupby("State")
gb.get_group("Alabama").set_index("Year").head()
```

```python
act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"): #name is the state, group is the dataframe
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["Low.2018"]].rename(columns = {"Low.2018": name})
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns = {'Low.2018': name}))
        
act_min_wage.head()
```
