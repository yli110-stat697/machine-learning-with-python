# titles and legends

```python
x = [1, 2, 3]
y1 = [5, 7, 4]
y2 = [10, 14, 8]

plt.plot(x, y1, label = 'line 1')
plt.plot(x, y2, label = 'line 2')
plt.title("Graph\nsubtitle")
plt.legend()

plt.xlabel("x")
plt.ylabel("y")
```

# bar charts and histograms

```python
x = list(range(2,11, 2))
x
```

```python
y = [1, 4, 5, 4, 3]
x2 = list(range(1,12, 2))
x2
```

```python
y2 = [3, 6, 7, 4, 4, 2]
```

```python
plt.bar(x, y, label = "bar chart 1")
plt.bar(x2, y2, label = "bar chart 2")

plt.xlabel("x")
plt.ylabel("y")
plt.title("bar charts\ncheck it out")
plt.legend()
```

```python
plt.bar(x, y, label = "bar chart 1", color = "r")
plt.bar(x2, y2, label = "bar chart 2", color = "c")

plt.xlabel("x")
plt.ylabel("y")
plt.title("bar charts\ncheck it out")
plt.legend()
```

```python
import numpy as np
ages = range(20, 50)
population_ages = np.random.choice(ages, 40)
ids = [i for i in range(len(population_ages))]
```

```python
plt.bar(population_ages, ids)

plt.xlabel("ids")
plt.ylabel("ages")
plt.title("bar charts for ages\nwho is the oldest?")
```

You can see that bar chart doesn't make sense. Histogram is better.

```python
bins = list(range(20, 51, 5))
bins
```

```python
plt.hist(population_ages, bins, histtype = "bar", rwidth = 0.8)
```

# stack plots

```python
days = list(range(1,6))
days
```

```python
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [8,8,7,6,5]
playing = [1,2,2,1,3]
```

```python
plt.stackplot(days, sleeping, eating, working, playing, colors = ['m','c','r','g'])
```

```python
plt.stackplot(days, sleeping, eating, working, playing, colors = ['m','c','r','g'])

plt.plot([], [], color = 'm', label = 'sleeping')
plt.plot([], [], color = 'c', label = 'eating')
plt.plot([], [], color = 'r', label = 'working')
plt.plot([], [], color = 'g', label = 'playing')

plt.legend()
```

# pie chart

```python
slices = [7,2,2,13]
activities = ['sleep', 'eat', 'work', 'play']
cols = ['m','c','g','r']

plt.pie(slices, labels = activities, colors = cols)
```

```python
plt.pie(slices, 
        labels = activities, 
        colors = cols, 
        shadow = True, 
        explode = (0, 0.1, 0, 0),
        autopct = '%1.1f%%')
```
